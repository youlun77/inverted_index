(3-2)jieba-cut

```python
import jieba
jieba.set_dictionary('dict.txt.big')
jieba.load_userdict("userdict.txt")

dict={}
for i in range(len(data)):
  words=jieba.cut(data[i]['articles']) #list1[]改articles
  #print(data[i]['articles'])
  for word in words:
    #print(word)
    #print(type(word))
    if word not in dict:
      dict[word]=[]
      dict[word].append(data[i]['id'])
    else:
      dict[word].append(data[i]['id'])
print(dict)
```

```python
#去掉重複document
for index in dict:
  unique_set=set(dict[index])
  dict[index]=list(unique_set)
#sort
for index in dict:
  list.sort(dict[index])
```
