```python
import jieba
jieba.set_dictionary('dict.txt.big')
jieba.load_userdict("userdict.txt")

dict={}
for i in range(len(data)):
  words=jieba.cut(data[i]['articles']) 
  for word in words:
    if word not in dict:
      dict[word]=[]
      dict[word].append(data[i]['id'])
    else:
      dict[word].append(data[i]['id'])
```

```python
#去掉重複document
for index in dict:
  unique_set=set(dict[index])
  dict[index]=list(unique_set)

#sort
for index in dict:
  dict[index].sort()
```
